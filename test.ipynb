```python


from pydub import AudioSegment
import noisereduce as nr
import matplotlib.pyplot as plt
import numpy as np

# Load audio file
sound = AudioSegment.from_file('./test2.wav', format="wav")

# Convert audio file to a numpy array
samples = np.array(sound.get_array_of_samples())

# Apply noise reduction
reduced_noise = nr.reduce_noise(y=samples, sr=sound.frame_rate)


# Plot the original and reduced audio signals
plt.subplot(2, 1, 1)
plt.title("Original Signal")
plt.plot(samples)

plt.subplot(2, 1, 2)
plt.title("Reduced Noise")
plt.plot(reduced_noise)

plt.tight_layout()
plt.show()
reduced_sound = AudioSegment(
    reduced_noise.tobytes(),
    frame_rate=sound.frame_rate,
    sample_width=sound.sample_width,
    channels=sound.channels
)

# Save the reduced noise audio to a file
reduced_sound.export("./reduced_noise.wav", format="wav")
```
